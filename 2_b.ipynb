```python
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as sn
from skimage import filters, color
import matplotlib.pylab as plt
import colorsys

# opening the image and converting it to grayscale
a1 = iio.imread('exercise/baby.jpeg')
a2 = iio.imread('exercise/balloons_noisy.png')
a3 = iio.imread('exercise/flower.jpeg')

# initializing the filter of size 5 by 5
# the filter is divided by 25 for normalization
k = np.ones((5, 5)) / 25

# performing convolution
if a1.ndim ==3:
    b1 = np.zeros_like(a1)
    for c in range(3):
        b1[:,:,c] = sn.convolve(a1[:,:,c], k).astype(np.uint8)
else:
    b1 = sn.convolve(a1, k).astype(np.uint8)

if a2.ndim ==3:
    b2 = np.zeros_like(a2)
    for c in range(3):
        b2[:,:,c] = sn.convolve(a2[:,:,c], k).astype(np.uint8)
else:
    b2 = sn.convolve(a2, k).astype(np.uint8)

if a3.ndim ==3:
    b3 = np.zeros_like(a3)
    for c in range(3):
        b3[:,:,c] = sn.convolve(a3[:,:,c], k).astype(np.uint8)
else:
    b3 = sn.convolve(a3, k).astype(np.uint8)

gray1 = color.rgb2gray(b1)
gray2 = color.rgb2gray(b2)
gray3 = color.rgb2gray(b3)


c1 = (filters.sobel(gray1)* 255).astype(np.uint8)
c2 = (filters.sobel(gray2)* 255).astype(np.uint8)
c3 = (filters.sobel(gray3)* 255).astype(np.uint8)

iio.imsave('baby_sobel_filter_edge_detection.png', c1)
iio.imsave('balloons_sobel_filter_edge_detection.png', c2)
iio.imsave('flower_sobel_filter_edge_detection.png', c3)

plt.subplot(1,3,1)
plt.imshow(c1)
plt.title("Baby (sobel)")


plt.subplot(1,3,2)
plt.imshow(c2)
plt.title("Balloons (sobel)")


plt.subplot(1,3,3)
plt.imshow(c3)
plt.title("Flower (sobel)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import imageio.v2 as iio
import scipy.ndimage as sn
from skimage import feature, color
import matplotlib.pylab as plt
import colorsys

# opening the image and converting it to grayscale
a1 = iio.imread('exercise/baby.jpeg')
a2 = iio.imread('exercise/balloons_noisy.png')
a3 = iio.imread('exercise/flower.jpeg')

# initializing the filter of size 5 by 5
# the filter is divided by 25 for normalization
k = np.ones((5, 5)) / 25

# performing convolution
if a1.ndim ==3:
    b1 = np.zeros_like(a1)
    for c in range(3):
        b1[:,:,c] = sn.convolve(a1[:,:,c], k).astype(np.uint8)
else:
    b1 = sn.convolve(a1, k).astype(np.uint8)

if a2.ndim ==3:
    b2 = np.zeros_like(a2)
    for c in range(3):
        b2[:,:,c] = sn.convolve(a2[:,:,c], k).astype(np.uint8)
else:
    b2 = sn.convolve(a2, k).astype(np.uint8)

if a3.ndim ==3:
    b3 = np.zeros_like(a3)
    for c in range(3):
        b3[:,:,c] = sn.convolve(a3[:,:,c], k).astype(np.uint8)
else:
    b3 = sn.convolve(a3, k).astype(np.uint8)

gray1 = color.rgb2gray(b1)
gray2 = color.rgb2gray(b2)
gray3 = color.rgb2gray(b3)


c1 = (feature.canny(gray1, sigma=3)* 255).astype(np.uint8)
c2 = (feature.canny(gray2, sigma=3)* 255).astype(np.uint8)
c3 = (feature.canny(gray3, sigma=3)* 255).astype(np.uint8)

iio.imsave('baby_canny_filter_edge_detection.png', c1)
iio.imsave('balloons_canny_filter_edge_detection.png', c2)
iio.imsave('flower_canny_filter_edge_detection.png', c3)

plt.subplot(1,3,1)
plt.imshow(c1)
plt.title("Baby (canny)")


plt.subplot(1,3,2)
plt.imshow(c2)
plt.title("Balloons (canny)")


plt.subplot(1,3,3)
plt.imshow(c3)
plt.title("Flower (canny)")

plt.tight_layout()
plt.show()
```
